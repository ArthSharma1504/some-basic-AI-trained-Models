```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Create the dataset
data = {
    'Hours_Studied': [1, 2, 3, 4, 5],
    'Score': [20, 40, 50, 65, 85]
}
df = pd.DataFrame(data)

# Step 2: Prepare data for model
X = df[['Hours_Studied']]  # features (2D)
y = df['Score']            # target (1D)

# Step 3: Train the model
model = LinearRegression()
model.fit(X, y)

# Step 4: Predict scores (including for plotting)
predicted_scores = model.predict(X)

# Step 5: Plot actual data and the regression line
plt.scatter(X, y, color='blue', label='Actual Scores')  # original points
plt.plot(X, predicted_scores, color='red', label='Regression Line')  # prediction line
plt.title('Hours Studied vs Score')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()

# Step 6: Predict for a new value
hours = 6
predicted_score = model.predict([[hours]])
print(f"\n🎯 If a student studies for {hours} hours, predicted score is: {predicted_score[0]:.2f}")
```
